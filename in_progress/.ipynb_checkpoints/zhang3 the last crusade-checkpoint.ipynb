```python
max_window_size = 5

def w_k(k, b = 2):
    return(2 * k * b + 1)

def slope(dhmax, k, w_k):
    pass

w_k_list = list(map(w_k, range(1, max_window_size+1)))
```

```python
# Set parameters
c = a_grid.cell_size
b = 2
dhmax = 2.5
m = a_grid.m
n= a_grid.n
```

```python
def erosion(Z, w_k):
    Z_f = []
    for j in range(1, n+1):
        # Construct vector of l indices
        lb = np.floor(1 - (w_k / 2))
        ub = np.ceil(1 + (w_k / 2))
        
        a = [Z[l] for l in range(int(lb), int(ub)) if l >=0 and l <= len(Z)-1]
        Z_f.append(np.min(a))
    return(Z_f)
    
def dilation(Z, w_k):
    Z_f = []
    for j in range(1, n+1):
        # Construct vector of l indices
        lb = np.floor(1 - (w_k / 2))
        ub = np.ceil(1 + (w_k / 2))
        
        a = [Z[l] for l in range(int(lb), int(ub)) if l >=0 and l <= len(Z)-1]
        Z_f.append(np.max(a))
    return(Z_f)
```

```python
for w_k in w_k_list:
    for i in range(1, m+1):
        P_i = A[i,:]
        Z = P_i
        Z_f = erosion(Z, w_k)
        #Z_f = dilation(Z_f, w_k)
        print(Z_f)
```
